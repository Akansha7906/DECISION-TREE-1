Question 1: What is a Decision Tree, and how does it work in the context of
classification?

•A Decision Tree is a supervised machine learning algorithm used for classification and regression tasks


•It works by splitting the dataset into branches based on feature values to make decisions


•Each internal node represents a feature, each branch represents a decision rule, and each leaf node represents a class label (for classification)


•The tree makes predictions by traversing from the root node to a leaf node according to the feature values of a given sample


•It uses metrics like Gini Index or Entropy (Information Gain) to decide the best splits


•Example: Predicting whether a customer will buy a product (yes/no) based on features like age, income, and browsing history.

Question 2: Explain the concepts of Gini Impurity and Entropy as impurity measures
How do they impact the splits in a Decision Tree?

•Gini Impurity measures the probability of misclassifying a randomly chosen element if it was randomly labeled according to the class distribution in a node


•Entropy measures the amount of disorder or uncertainty in the class distribution of a node


•Both are impurity measures used to decide the best feature and threshold for splitting nodes in a Decision Tree


•A lower Gini or Entropy after a split indicates a purer node, meaning the samples are more homogenous


•Decision Trees choose the split that maximizes information gain (reduction in Entropy) or minimizes Gini Impurity, resulting in better classification performance.


Question 3: What is the difference between Pre-Pruning and Post-Pruning in Decision
Trees?

 Give one practical advantage of using each?

 •Pre-Pruning stops the tree from growing too deep by setting limits like maximum depth, minimum samples per leaf, or minimum information gain


•Post-Pruning first allows the tree to grow fully and then removes branches that do not improve performance on a validation set


•Advantage of Pre-Pruning: Reduces overfitting early and saves computation


•Advantage of Post-Pruning: More accurate final tree, as it considers the full structure before pruning.


Question 4: What is Information Gain in Decision Trees, and why is it important for
choosing the best split?

•Information Gain measures the reduction in entropy or uncertainty after splitting a node based on a feature


•It calculates how much useful information a feature provides about the target variable


•A higher Information Gain means the feature better separates the classes, leading to a purer node


•Decision Trees use Information Gain to choose the best feature and threshold for splitting at each node


•It is important because it ensures each split improves classification accuracy and builds an effective tree.

Question 5: What are some common real-world applications of Decision Trees, and
what are their main advantages and limitations?

•Decision Trees are used in customer churn prediction, credit scoring, fraud detection, medical diagnosis, and marketing targeting


**•Advantages**:
•Easy to understand and interpret
•Can handle both numerical and categorical data


•Requires little data preprocessing


**•Limitations**:
•Prone to overfitting, especially with deep trees.
•Can be unstable, small changes in data may lead to a different tree.
•May be biased toward features with more levels
